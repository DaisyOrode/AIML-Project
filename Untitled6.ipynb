Understanding the Structure of the Fiscal Dataset

Before diving into modeling and insights, it is crucial to thoroughly understand the structure and composition of our fiscal dataset. The following code block performs a structured exploration of the data to reveal its characteristics and potential quality issues.

```python
# ============================
# 7️⃣ NORMALIZE ALL AMOUNTS TO BILLIONS
# ============================

def to_billion(unit, amount):
    """
    Convert fiscal amounts to a standardized Billions scale.
    Handles Million vs Billion vs non-monetary units.
    """
    u = str(unit).lower()

    if pd.isna(amount):
        return np.nan

    # Convert Millions → Billions
    if "million" in u:
        return amount / 1000

    # Keep Billions as is
    if "billion" in u:
        return amount

    # Non-monetary (%, points, persons) → keep raw
    return amount

df["Amount_Billion"] = df.apply(lambda row: to_billion(row["Unit"], row["Amount"]), axis=1)

# Preview the normalized column
df[["Country", "Indicator", "Unit", "Amount", "Amount_Billion"]].head()
```

Normalizing Fiscal Data into a Common Scale

To conduct accurate fiscal analysis — particularly for macroeconomic indicators such as Government Debt, Nominal GDP, Revenue, and Expenditure — we must ensure that all monetary values are expressed in the same unit of measurement.

Auditing Reporting Frequencies Across the Dataset

One of the most common challenges in national fiscal datasets is mixed reporting frequencies.
Different indicators — and sometimes the same indicator within a country — may be reported:

Monthly

Quarterly

Annually

This inconsistency can lead to severe analytical issues such as:

Duplicate values for the same year

Artificial spikes in computed ratios

Incorrect aggregation

Misleading time-series patterns

To address this, we performed a frequency audit.

Why This Step Matters

Understanding frequency inconsistencies is crucial because mixed monthly, quarterly, and yearly reports cause duplicate time entries and lead to misleading trends and ratios. This audit highlights which indicators need frequency harmonization before we can build reliable annual fiscal metrics.

```python
# ============================
# 9️⃣ SELECT PRIMARY SOURCE PER COUNTRY–INDICATOR
# ============================

# 1. Count how many rows we have for each Country–Indicator–Source
source_audit = (
    df.groupby(['Country', 'Indicator', 'Source'])
      .size()
      .reset_index(name='Row_Count')
      .sort_values(['Country', 'Indicator', 'Row_Count'], ascending=[True, True, False])
)

print("=== Sample of Source Audit ===")
print(source_audit.head(20))

# 2. For each Country–Indicator, select the Source with the most rows
primary_sources = (
    source_audit
    .groupby(['Country', 'Indicator'])
    .first()             # after sorting, 'first' = highest Row_Count
    .reset_index()
)

print("\n=== Primary Source Selected per Country–Indicator ===")
print(primary_sources.head(20))

# 3. Filter the main dataframe to keep only the primary source for each Country–Indicator
df_primary = df.merge(
    primary_sources[['Country', 'Indicator', 'Source']],
    on=['Country', 'Indicator', 'Source'],
    how='inner'
)

print("\n=== Shape Before vs After Source Filtering ===")
print("Original rows:", len(df))
print("Filtered rows:", len(df_primary))

# Quick preview
df_primary[['Country', 'Indicator', 'Source', 'Frequency', 'Unit']].head()
```

Choosing a Consistent Data Source for Each Indicator

In real-world fiscal datasets, the same indicator can be reported by multiple institutions — for example, the Central Bank, the National Statistical Office, and the Ministry of Finance may all publish debt or GDP figures. If we mix these series without control, we introduce inconsistency and artificial volatility.

To avoid this, we enforce a single primary source for each country and indicator.

Why This Step Matters (Good for Slides)

“Multiple institutions often report the same indicator using different methodologies. To prevent series mixing and artificial volatility, we standardized each country’s indicator to a single primary source — chosen as the source with the most complete data. This ensures that our time series are consistent, interpretable, and suitable for constructing ratios and trends such as Debt-to-GDP.

```python
# ============================
# 🔟 AGGREGATE TO ANNUAL VALUES (CLEAN PANEL)
# ============================

# Define how each indicator behaves
indicator_types = {
    "Government Debt": "stock",
    "Nominal GDP": "stock",
    "Real GDP": "stock",
    "Budget Deficit/Surplus": "flow",
    "Revenue": "flow",
    "Expenditure": "flow",
    "Tax Revenue": "flow",
    "Inflation Rate": "other",
    "GDP Growth Rate": "other"
}

df_primary_key = df_primary[df_primary['Indicator'].isin(indicator_types.keys())].copy()

# Annual aggregator function
def annual_agg(group):
    indicator = group.name[1]
    kind = indicator_types.get(indicator, "stock")

    # If Yearly frequency exists, use it directly
    yearly = group[group['Frequency'] == 'Yearly']
    if not yearly.empty:
        val_billion = yearly['Amount_Billion'].median()
        val_raw = yearly['Amount'].median()
    else:
        # Otherwise aggregate lower frequencies
        if kind == "flow":
            val_billion = group['Amount_Billion'].sum()      # flows summed annually
        elif kind == "stock":
            val_billion = group['Amount_Billion'].median()   # stocks use median
        else:
            val_billion = group['Amount'].mean()             # percentages/indices use avg
        val_raw = group['Amount'].mean()

    return pd.Series({
        "Annual_Value_Billion": val_billion,
        "Annual_Value_Raw": val_raw
    })

# Apply the aggregator
annual = (
    df_primary_key
    .groupby(['Country','Indicator','Year'], as_index=False)
    .apply(annual_agg)
)

# Choose which field to use depending on indicator type
monetary_inds = [k for k,v in indicator_types.items() if v in ("stock","flow")]

annual['Annual_Value'] = np.where(
    annual['Indicator'].isin(monetary_inds),
    annual['Annual_Value_Billion'],
    annual['Annual_Value_Raw']
)

# Create wide (pivoted) annual panel
panel = annual.pivot_table(
    index=['Country','Year'],
    columns='Indicator',
    values='Annual_Value',
    aggfunc='mean'
).reset_index()

panel.columns.name = None

print("=== Clean Annual Panel Preview ===")
panel.head()
```

Creating a Clean Annual Fiscal Dataset

Once I selected the primary source for each indicator and country, the next step is to harmonize reporting frequencies. Many indicators are reported Monthly, Quarterly, or Yearly, and combining these directly leads to distorted insights.

This code block aggregates all fiscal indicators into a single annual value, creating a clean panel dataset that is ready for accurate trend analysis and ratio calculations.

I transformed all fiscal indicators into harmonized annual values, accounting for different reporting frequencies and economic behavior of each indicator. This produces a clean, consistent panel dataset across all countries, forming the foundation for accurate fiscal ratios and insight generation.”

```python
# ============================
# 1️⃣1️⃣ COMPUTE FISCAL RATIOS
# ============================

# Ensure required columns exist
required_cols = [
    "Government Debt", "Nominal GDP", "Budget Deficit/Surplus",
    "Revenue", "Expenditure", "Tax Revenue"
]

print("Checking required columns...\n")
for col in required_cols:
    print(col, "→", "OK" if col in panel.columns else "MISSING")

# Calculate ratios (all expressed as percentages of GDP)
panel['Debt_to_GDP'] = (panel['Government Debt'] / panel['Nominal GDP']) * 100
panel['Deficit_to_GDP'] = (panel['Budget Deficit/Surplus'] / panel['Nominal GDP']) * 100
panel['Revenue_to_GDP'] = (panel['Revenue'] / panel['Nominal GDP']) * 100
panel['Expenditure_to_GDP'] = (panel['Expenditure'] / panel['Nominal GDP']) * 100
panel['Tax_to_GDP'] = (panel['Tax Revenue'] / panel['Nominal GDP']) * 100

# Preview only meaningful columns
ratio_view = panel[['Country', 'Year',
                    'Debt_to_GDP', 'Deficit_to_GDP',
                    'Revenue_to_GDP', 'Expenditure_to_GDP', 'Tax_to_GDP']]

print("\n=== Fiscal Ratios Preview ===")
ratio_view.head()
```

```python
# ============================
# 🌐 FISCAL RADAR CHART (PER COUNTRY)
# ============================

import numpy as np
import matplotlib.pyplot as plt

def plot_fiscal_radar(country):
    """
    Plot a radar chart for one country's latest fiscal ratios:
    - Deficit_to_GDP
    - Revenue_to_GDP
    - Expenditure_to_GDP
    - Tax_to_GDP
    """

    # Pick latest year for this country
    temp = panel[panel['Country'] == country].sort_values('Year')
    if temp.empty:
        print(f"No data available for {country}.")
        return

    latest_row = temp.iloc[-1]  # latest year

    # Metrics and labels
    metrics = ['Deficit_to_GDP', 'Revenue_to_GDP', 'Expenditure_to_GDP', 'Tax_to_GDP']
    labels = ['Deficit-to-GDP', 'Revenue-to-GDP', 'Expenditure-to-GDP', 'Tax-to-GDP']

    values = [latest_row[m] for m in metrics]

    # Handle all-NaN or too many NaNs safely
    if all(pd.isna(values)):
        print(f"Fiscal ratios missing for {country} (latest year).")
        return

    # Replace NaNs with 0 for plotting, but you could choose another strategy
    values = [0 if pd.isna(v) else v for v in values]

    # Close the loop by repeating the first value
    values += values[:1]

    # Angles for each axis
    num_vars = len(metrics)
    angles = np.linspace(0, 2 * np.pi, num_vars + 1)

    # Create polar plot
    fig, ax = plt.subplots(figsize=(7, 7), subplot_kw=dict(polar=True))

    # Plot and fill
    ax.plot(angles, values, linewidth=2, linestyle='solid')
    ax.fill(angles, values, alpha=0.25)

    # Set category labels
    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(labels, fontsize=10)

    # Add title with country and year
    year = int(latest_row['Year'])
    ax.set_title(f"{country} – Fiscal Ratios Radar ({year})", fontsize=14, fontweight='bold', pad=20)

    # Optional: tweak radial limits automatically or set manually
    # ax.set_ylim(bottom=min(0, min(values)), top=max(values) * 1.2)

    plt.show()


#plot_fiscal_radar("South Africa")
plot_fiscal_radar("Nigeria")
# plot_fiscal_radar("Ghana")
# plot_fiscal_radar("Egypt")
```

```python
# ============================
# 1️⃣2️⃣ MODEL 2 — FISCAL RISK CLASSIFICATION (KMEANS)
# ============================

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import numpy as np

# 1. Take the most recent year available for each country
latest = (
    panel.sort_values('Year')
          .groupby('Country')
          .tail(1)     # last row per country = latest year
          .reset_index(drop=True)
)

print("=== Latest Fiscal Snapshot per Country ===")
print(latest[['Country', 'Year',
              'Debt_to_GDP','Deficit_to_GDP',
              'Revenue_to_GDP','Expenditure_to_GDP',
              'Tax_to_GDP','GDP Growth Rate','Inflation Rate']])

# 2. Select features for risk classification
features = [
    'Debt_to_GDP',
    'Deficit_to_GDP',
    'Revenue_to_GDP',
    'Expenditure_to_GDP',
    'Tax_to_GDP',
    'GDP Growth Rate',
    'Inflation Rate'
]

X = latest[features].copy()

# Handle any remaining missing values by simple imputation (e.g., column mean)
X_imputed = X.fillna(X.mean())

# 3. Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_imputed)

# 4. Apply KMeans clustering to group countries into 3 risk tiers
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
clusters = kmeans.fit_predict(X_scaled)

latest['Risk_Cluster'] = clusters

# 5. Interpret clusters: rank them by average Debt_to_GDP and Deficit_to_GDP
cluster_summary = (
    latest.groupby('Risk_Cluster')[['Debt_to_GDP','Deficit_to_GDP']]
          .mean()
          .sort_values('Debt_to_GDP', ascending=True)
)

print("\n=== Cluster Summary (Lower to Higher Risk) ===")
print(cluster_summary)

# Map cluster ids to human-readable labels based on their ranking
ordered_clusters = cluster_summary.index.tolist()  # [low_risk_cluster_id, mid_risk_cluster_id, high_risk_cluster_id]

cluster_to_label = {
    ordered_clusters[0]: "Low Risk",
    ordered_clusters[1]: "Medium Risk",
    ordered_clusters[2]: "High Risk"
}

latest['Fiscal_Risk_Level'] = latest['Risk_Cluster'].map(cluster_to_label)

# 6. Final risk table
risk_table = latest[['Country', 'Year',
                     'Debt_to_GDP','Deficit_to_GDP',
                     'Revenue_to_GDP','Expenditure_to_GDP','Tax_to_GDP',
                     'GDP Growth Rate','Inflation Rate',
                     'Fiscal_Risk_Level']].sort_values('Fiscal_Risk_Level')

print("\n=== Fiscal Risk Classification ===")
print(risk_table)

# 7. Optional: save for Power BI / report
risk_table.to_csv("Fiscal_Risk_Classification_Latest.csv", index=False)
```

Data-Driven Fiscal Risk Classification

To move beyond simple descriptive statistics, I built a Fiscal Risk Classification model that groups African countries into Low, Medium, and High fiscal risk based on their latest macro-fiscal conditions.

I use KMeans clustering to let the data itself reveal natural groupings of risk levels

```python
# ============================
# SIMPLE MODEL 2 — FISCAL RISK CLASSIFICATION
# ============================

# 1. Take the most recent year for each country
latest = (
    panel.sort_values('Year')
         .groupby('Country')
         .tail(1)
         .reset_index(drop=True)
)

# 2. Define a simple rule-based classifier
def classify_fiscal_risk(row):
    debt = row['Debt_to_GDP']
    deficit = row['Deficit_to_GDP']  # negative = deficit

    # Handle missing values safely
    if pd.isna(debt) and pd.isna(deficit):
        return "Data Gap"

    # HIGH RISK: very high debt OR large deficits
    if (not pd.isna(debt) and debt > 80) or (not pd.isna(deficit) and deficit < -5):
        return "High Risk"

    # MEDIUM RISK: moderately high debt OR moderate deficits
    if (not pd.isna(debt) and debt > 60) or (not pd.isna(deficit) and deficit < -3):
        return "Medium Risk"

    # Otherwise: relatively safe for now
    return "Low Risk"

# 3. Apply classification
latest['Fiscal_Risk_Level'] = latest.apply(classify_fiscal_risk, axis=1)

# 4. Build a clean output table
risk_table = latest[['Country', 'Year',
                     'Debt_to_GDP', 'Deficit_to_GDP',
                     'Revenue_to_GDP', 'Expenditure_to_GDP', 'Tax_to_GDP',
                     'Fiscal_Risk_Level']].sort_values('Fiscal_Risk_Level')

print("=== Fiscal Risk Classification (Rule-Based) ===")
print(risk_table)

# 5. Optional: quick summary count
print("\n=== Count by Risk Level ===")
print(risk_table['Fiscal_Risk_Level'].value_counts())

# 6. Save for dashboard / slides
risk_table.to_csv("Fiscal_Risk_Classification_Simple.csv", index=False)
```

Simple, Transparent Fiscal Risk Model

To keep the model easy to explain and defend in a hackathon setting, I used a rule-based fiscal risk classification grounded in two widely used indicators:

Debt-to-GDP – measures how large a country’s debt stock is relative to its economy.

Deficit-to-GDP – measures how much more the government spends than it earns each year.

Rather than using a complex black-box model, we apply clear threshold rules that any policymaker would recognize.

```python
# ============================
#  DEBT-TO-GDP FORECASTING (ARIMA)
# ============================

from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd # Import pandas for PeriodIndex

def forecast_debt(country, years_ahead=5):
    # 1. Select the country’s data
    temp = panel[panel['Country'] == country].sort_values('Year')
    ts = temp[['Year', 'Debt_to_GDP']].dropna()

    # Safety check
    if len(ts) < 5:
        print(f"Not enough data to forecast for {country}. Need at least 5 points.")
        return

    # Ensure 'Year' is an integer before setting as index
    ts['Year'] = ts['Year'].astype(int)
    ts = ts.set_index('Year')

    # Convert the index to a PeriodIndex for statsmodels
    # Using 'Y' for annual frequency as recommended by FutureWarning
    ts.index = pd.PeriodIndex(ts.index, freq='Y')

    # 2. Fit ARIMA model (simple and stable)
    model = ARIMA(ts['Debt_to_GDP'], order=(1,1,1))
    model_fit = model.fit()

    # 3. Forecast into the future
    forecast = model_fit.forecast(steps=years_ahead)

    # 4. Build a combined dataframe
    # Extract years from the forecast index which will be a PeriodIndex
    future_years = [p.year for p in forecast.index]
    forecast_df = pd.DataFrame({'Year': future_years, 'Forecast_Debt_to_GDP': forecast.values})

    # 5. Plot
    plt.figure(figsize=(12,5))

    # Prepare historical data for plotting: ensure 'Year' column is int
    plot_ts_historical = ts.reset_index()
    plot_ts_historical['Year'] = plot_ts_historical['Year'].apply(lambda p: p.year)

    sns.lineplot(x='Year', y='Debt_to_GDP', data=plot_ts_historical, marker='o', label='Historical')
    sns.lineplot(x='Year', y='Forecast_Debt_to_GDP', data=forecast_df, marker='o', label='Forecast')

    plt.title(f"{country} — Debt-to-GDP Historical Trend and {years_ahead}-Year Forecast")
    plt.ylabel("Debt-to-GDP (%)")
    plt.xlabel("Year")
    plt.grid(True)
    plt.legend()
    plt.show()

    return forecast_df

# Run the model for a country of your choice:
forecast_debt("South Africa")
```

```python
forecast_debt("Ghana")
forecast_debt("Egypt")
forecast_debt("Nigeria")
forecast_debt("Kenya")
```
